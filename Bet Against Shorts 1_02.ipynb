```python
import alpaca_trade_api as tradeapi
import pandas as pd
import numpy as np
from IPython.display import clear_output
import timeit
import finviz as fv
from datetime import datetime
import os
```

```python
APCA_API_BASE_URL = 'https://paper-api.alpaca.markets'
APCA_API_KEY_ID = '###'
APCA_API_SECRET_KEY = '###'
api = tradeapi.REST(APCA_API_KEY_ID,APCA_API_SECRET_KEY,APCA_API_BASE_URL)
```

```python
active_assets=api.list_assets(status='active')
symbols = [a.symbol for a in active_assets]
len(symbols)
```

```python
fv_fails = []
fv_df = pd.DataFrame()
c = 0
start = timeit.default_timer()
for i in symbols:
    stop = timeit.default_timer()
    time_elapsed = stop-start
    minutes_elapsed = np.round((time_elapsed/60),decimals=2)
    percent = np.round((c)/len(symbols),decimals=5)*100
    expected_time = np.round((minutes_elapsed/percent*100),decimals=2)
    clear_output(wait=True)
    try:
        data = fv.get_stock(i)
        data_df = pd.DataFrame(data,index=[0])
        data_df['Symbol'] = i
        fv_df = fv_df.append(data_df,ignore_index=True)
        c += 1
        print(c,'of',len(symbols),'-',percent,'%',i,)
        print('Runtime so far:',minutes_elapsed,'minutes')
        print('Expected runtime:',expected_time,'minutes')
        print(len(fv_fails),'Fails:',fv_fails)
    except:
        fv_fails.append(i)
        print(c,'of',len(symbols),'-',percent,'%',i,)
        print('Runtime so far:',minutes_elapsed,'minutes')
        print('Expected runtime:',expected_time,'minutes')
        print(len(fv_fails),'Fails:',fv_fails)
        pass
fv_df.set_index('Symbol',inplace=True)
df = fv_df
```

```python
fv_df = pd.read_csv('clean_finviz_extract23-04-2020_12-02-33_AM.csv')
fv_df.set_index('Symbol',inplace=True)
df = fv_df
df.head()
```

```python
KMB_cols = ['Shs Outstand', 'Shs Float','Market Cap','Income','Sales']
for col in KMB_cols:
    try:
        df[col] = df[col].replace(r'[KMB]+$', '', regex=True).astype(float) * df[col].str.extract(r'[\d\.]+([KMB]+)', expand=False).fillna(1).replace(['K','M','B'], [10**3, 10**6, 10**9]).astype(int)
    except:
        pass
pct_cols = ['Short Ratio', 'Short Float', 'Insider Own','Insider Trans','Inst Own', 'Inst Trans','ROI','ROE','ROA']
for col in pct_cols:
    try:
        df[col] = df[col].replace('%','*.01',regex=True).dropna().map(pd.eval).astype(float)
    except:
        pass
df.head()
now = datetime.now().strftime("%d-%m-%Y_%I-%M-%S_%p")
filename = "clean_finviz_extract"+now+".csv"
df.to_csv(filename)
```

```python
ma_length = 10
df.MADev = 0
df.OneDayReturn = 0
df.DollarVol = 0
df.DollarVolLiquidity = 0
df.TotalRollingDollarVol = 0
fails = []
c = 0
start = timeit.default_timer()
for i in df.index:
    stop = timeit.default_timer()
    time_elapsed = stop-start
    minutes_elapsed = np.round((time_elapsed/60),decimals=2)
    percent = np.round((c)/len(df.index),decimals=5)*100
    expected_time = np.round((minutes_elapsed/percent*100),decimals=2)
    try:
        clear_output(wait=True)
        c+= 1
        bars = api.get_barset(i,'day',limit=ma_length).df
        bars = bars[i]
        bars.OneDayReturn = bars['close'].pct_change()
        bars.MA = bars['close'].rolling(window=ma_length).mean()
        bars.MADev = (bars['close'] - bars.MA) /  bars.MA
        bars.DollarVol = bars['close'] * bars['volume']
        bars.TotalRollingDollarVol= bars.DollarVol.rolling(window=ma_length).sum()
        bars.DollarVolLiquidity = bars.DollarVol / bars.TotalRollingDollarVol
        df.at[i,'MADev'] = bars.MADev[-1]
        df.at[i,'OneDayReturn'] = bars.OneDayReturn[-1]
        df.at[i,'DollarVol'] = bars.DollarVol[-1]
        df.at[i,'DollarVolLiquidity'] = bars.DollarVolLiquidity[-1]
        df.at[i,'TotalRollingDollarVol'] = bars.TotalRollingDollarVol[-1]
        print(c,'of',len(df.index),'-',percent,'%',i,)
        print('Runtime so far:',minutes_elapsed,'minutes')
        print('Expected runtime:',expected_time,'minutes')
        print(len(fails),'Fails:',fails)
    except:
        fails.append(i)
        print(c,'of',len(df.index),i, '-FAIL')
        pass
```

```python
df2 = df[df['Shs Float'].notnull()]
cols_to_rank = {'MADev':True,'DollarVolLiquidity':True,'Short Ratio':False}
df2['AggRank'] = 0
for key,value in cols_to_rank.items():
    name = key + 'Rank'
    df2[name] = df2[key].rank(ascending=value)
    df2['AggRank'] += df2[name]
```

```python
pipe = df2[['AggRank','MADev','MADevRank','DollarVolLiquidity','DollarVolLiquidityRank','Short Ratio','Short RatioRank']].sort_values(by='AggRank').head(20)
```

```python
c = 0
for i in pipe.index:
    bars = api.get_barset(i,'day',limit=2)
    bars = bars[i]
    prior_period = bars[0].c
    prior_volume = bars[0].v
    current_period = bars[1].c
    current_volume = bars[1].v
    current_open = bars[1].o
    current_high = bars[1].h
    current_low = bars[1].l
    oc_pct = (current_period - current_open) / current_open *100
    pct_chg = (current_period - prior_period) / prior_period * 100
    pipe.at[i,'TodayChange'] = pct_chg
    pipe.at[i,'YesterdayClose'] = prior_period
    pipe.at[i,'TodayVolume'] = current_volume
    pipe.at[i,'TodaySpot'] = current_period
    pipe.at[i,'TodayOpen'] = current_open
    pipe.at[i,'TodayHigh'] = current_high
    pipe.at[i,'TodayLow'] = current_low
    pipe.at[i,'OpenClose%Chg'] = oc_pct
    c += 1
    print(c, "of ",len(pipe.index)," - ",i," - ",pct_chg)
pipe
```

```python
account = api.get_account()
account
positions = 20
default_stop = .95
equity = float(account.equity)
LotDollarSize = equity/positions
LotDollarSize
```

```python
order_fails = []
for i in pipe.index[:positions]:
    quote = api.polygon.last_quote(i)
    order_price = ((quote.askprice + quote.bidprice)/2)
    stop_price = order_price * default_stop
    take_profit = order_price *1.05
    quantity = round(LotDollarSize/order_price)
    try:
        api.submit_order(symbol=i,side='buy',type='limit',limit_price=order_price,qty=quantity,time_in_force='day',order_class='bracket',take_profit=dict(limit_price=take_profit,),stop_loss=dict(stop_price=stop_price,))
        print(i,"- BUY -",order_price,'- QTY:',quantity,'// STOP - ',stop_price, quote)
    except:
        order_fails.append(i)
        pass
```
